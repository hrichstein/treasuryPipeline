```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f606w = np.genfromtxt('dir2503/HOR-I-cut-std_pix_matched_F606W.dat',names=True)
f814w = np.genfromtxt('dir2503/HOR-I-cut-std_pix_matched_F814W.dat',names=True)
```

```python
old_file = '/Users/hr8jz/Box Sync/Research/source_lists/june13/HOROLOGIUM-I_sfErr.dat'
```

```python
old_flc = '/Volumes/Spare Data/Hannah_Data/hor1dir6pix/HORI_pix_2212_d6_dist.dat'
```

```python
data = np.genfromtxt(old_file,names=True)
```

```python
flc = np.genfromtxt(old_flc,names=True)
```

```python
print(len(data))
print(len(flc))
print(len(f606w))
```

```python
fig, ax = plt.subplots(figsize=(7,9))

ax.scatter(f606w['mean']-f814w['mean'],f606w['mean'],label='Cut')
ax.scatter(data['magRaw_v']-data['magRaw_i'],data['magRaw_v'],label='DRC')
ax.scatter(flc['mean_f606w']-flc['mean_f814w'],flc['mean_f606w'],label='pastFLC')

ax.set_xlim(-2,2)
ax.set_ylim(28,18)
ax.legend()
plt.show()
```

```python
fig,ax = plt.subplots()

ax.scatter(f606w['mean'],f606w['stdev'],label='CR cleaned')
ax.scatter(flc['mean_f606w'],flc['stdev_f606w'],label='FLCs, not cleaned',color='pink')

ax.set_xlim(18,28)
ax.set_ylim(-0.01,0.5)
ax.legend()
plt.show()
```

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(15,5),sharey=True,sharex=True)

ax1.scatter(f606w['mag1']-f814w['mag1'],f606w['mag1'],label='CR flc1')
ax2.scatter(f606w['mag2']-f814w['mag2'],f606w['mag2'],label='CR flc2')
ax3.scatter(f606w['mag3']-f814w['mag3'],f606w['mag3'],label='CR flc3')
ax4.scatter(f606w['mag4']-f814w['mag4'],f606w['mag4'],label='CR flc4')

ax1.scatter(flc['mag1_f606w']-flc['mag1_f814w'],flc['mag1_f606w'],label='First flc1')
ax2.scatter(flc['mag2_f606w']-flc['mag2_f814w'],flc['mag2_f606w'],label='First flc2')
ax3.scatter(flc['mag3_f606w']-flc['mag3_f814w'],flc['mag3_f606w'],label='First flc3')
ax4.scatter(flc['mag4_f606w']-flc['mag4_f814w'],flc['mag4_f606w'],label='First flc4')

ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()

ax1.set_ylim(28,18)
ax1.set_xlim(-2,2)

plt.show()
```
