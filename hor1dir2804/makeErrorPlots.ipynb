```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
new_cat = np.genfromtxt('HORI_comb0605.dat',names=True)
psf = np.genfromtxt('../elena_cats/HOROLOGIUM_CF.1.PSF.CAT',names=True)
old_cat = np.genfromtxt('../magsWerrs0811/HORI_final_0612.dat',names=True)
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,sharex=True,sharey=True,figsize=(12,8))

ax1.scatter(old_cat['median_f606w'],old_cat['stdev_f606w'],label='Last Catalog',s=5,color='sandybrown')
ax1.scatter(new_cat['zptMean_f606w'],new_cat['zptErr_f606w'],label='New Catalog',s=5,color='darkmagenta')

ax3.scatter(old_cat['median_f814w'],old_cat['stdev_f814w'],label='Last Catalog',s=5,color='sandybrown')
ax3.scatter(new_cat['zptMean_f814w'],new_cat['zptErr_f814w'],label='New Catalog',s=5,color='darkmagenta')

ax2.scatter(psf['m606c'],psf['s606'],label='PSF',s=5,color='lightcoral')
ax2.scatter(new_cat['zptMean_f606w'],new_cat['zptErr_f606w'],label='New Catalog',s=5,color='darkmagenta')

ax4.scatter(psf['m606c'],psf['s606'],label='PSF',s=5,color='lightcoral')
ax4.scatter(new_cat['zptMean_f606w'],new_cat['zptErr_f606w'],label='New Catalog',s=5,color='darkmagenta')


ax1.set_ylabel('F606W Err',fontsize=14)
ax3.set_ylabel('F814W Err',fontsize=14)

ax1.set_xlabel('F606W',fontsize=14)
ax2.set_xlabel('F606W',fontsize=14)

ax3.set_xlabel('F814W',fontsize=14)
ax4.set_xlabel('F814W',fontsize=14)

ax1.set_xlim(18,27.5)
ax1.set_ylim(0,0.55)

ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()

plt.subplots_adjust(wspace=0)

plt.savefig('newErrorPlots.png')
# plt.show()
```
