```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f606w = np.genfromtxt('HOR-I-cut-std_pix_matched_F606W_nC2.dat',names=True)
f814w = np.genfromtxt('HOR-I-cut-std_pix_matched_F814W_nC2.dat',names=True)
```

```python
f606w_cut = np.genfromtxt('../hor1dir1803/dir2503/HOR-I-cut-std_pix_matched_F606W.dat',names=True)
f814w_cut = np.genfromtxt('../hor1dir1803/dir2503/HOR-I-cut-std_pix_matched_F814W.dat',names=True)
```

```python
fig,ax = plt.subplots(figsize=(7,9))

ax.scatter(f606w['mean']-f814w['mean'],f606w['mean'],label='Uncut')
ax.scatter(f606w_cut['mean']-f814w_cut['mean'],f606w_cut['mean'],label='Sigma-clipped')

ax.set_xlim(-2,2)
ax.set_ylim(28,18)
ax.set_ylabel('F606W')
ax.set_xlabel('F606W-F814W')
ax.legend()

plt.show()
```

```python
fig,ax = plt.subplots()

ax.scatter(f606w['mean'],f606w['stdev'],label='Not cut')
ax.scatter(f606w_cut['mean'],f606w_cut['stdev'],label='cut')


ax.set_xlim(18,28)
ax.set_ylim(-0.01,0.5)
ax.legend()
plt.show()
```
